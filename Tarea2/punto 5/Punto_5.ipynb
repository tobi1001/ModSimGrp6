```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución triangular
a = 1
b = 10 # Rango
med = 4 # Media


# Genera silueta de la distribución deseada
def pdf(x):
  if a<= x < med: return 2*(x-a) / ((b-a)*(med-a))
  elif med <= x <= b: return 2*(b-x) / ((b-a)*(b-med))

  return 0

# Función para generar valores con distribución triangular
def inv_triangular():
  U = random.uniform(0, 1)
  if U < (med - a) / (b - a):
      x = a + (U * (b - a) * (med - a)) ** 0.5
  else:
      x = b - ((1 - U) * (b - a) * (b - med)) ** 0.5
  return x


# Genera una muestra de valores con distribución triangular
muestra = [inv_triangular() for i in range(60000)]

x = np.linspace( min(muestra), max(muestra), 500)
y = [ pdf(e) for e in x ]

# Grafica un histograma de la muestra
plt.hist(muestra, bins= 100, density=True, alpha=0.6, color='darkgreen', label='Muestra')
plt.plot(x, y, color = 'r')

plt.xlabel('Valor'); plt.ylabel('Frecuencia')
plt.title('Distribución Triangular')

plt.legend(); plt.show()
```
