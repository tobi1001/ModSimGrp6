```python
import numpy as np
import matplotlib.pyplot as plt

def inverse_cdf(u):
    if u <= 0.5:
        return -3 + 6 * (u - 0.5)
    else:
        return np.sqrt(32 * (u - 0.5))

def generate_samples(n):
    samples = []
    for _ in range(n):
        u = np.random.uniform(0, 1)
        samples.append(inverse_cdf(u))
    return samples

n_samples = 1000000
samples = generate_samples(n_samples)
plt.hist(samples, bins=50, density=True, alpha=0.6, color='g', edgecolor='black')
x = np.linspace(-3, 4, 1000)
pdf = np.piecewise(x, [x <= -3, (x > -3) & (x <= 0), (x > 0) & (x <= 4), x > 4],
                   [1/6, 0, lambda x: x/16,0])
plt.plot(x, pdf, 'r', linewidth=2)
plt.title('Funci√≥n de densidad de probabilidad')
plt.xlabel('x')
plt.ylabel('PDF')
plt.legend(['PDF', 'Histograma'])
plt.grid(True)
plt.show()
```
