```python
import numpy as np
import matplotlib.pyplot as plt

def pdf(x):
    return x**2 / 9

def inverse_cdf(u):
    return np.cbrt(27 * u)

def generate_random_numbers(n):
    random_numbers = []
    for _ in range(n):
        u = np.random.rand()
        x = inverse_cdf(u)
        random_numbers.append(x)
    return random_numbers

# Generar números aleatorios
n = 1000000
random_numbers = generate_random_numbers(n)

# Función de densidad de probabilidad teórica
x_values = np.linspace(0, 3, 100)
y_values = pdf(x_values)

plt.hist(random_numbers, bins=50, density=True, color='skyblue', edgecolor='black', label='Histograma de números aleatorios')
plt.plot(x_values, y_values, color='red', lw=2, label='PDF teórica')
plt.title('Función de densidad generada y PDF teórica')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()

```
