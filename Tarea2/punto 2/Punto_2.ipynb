### **Punto 2 a**

```python
import numpy as np
import random as r
import matplotlib.pyplot as plt

def pdf_display(x):
    # returning the value in the original PDF definition
    if 0 <= x <= 1:
        return 3/4
    elif 1<x<=5/3:
        return 9*(5/3 - x)/8
    else:
        return 0

def random_number_generation():
    u1 = r.uniform(0, 1)
    u2 = r.uniform(0, 1)
    Y = (5/3)*u2
    if Y <= 1:
        pa = 1
    else:
        pa = (5/3 - Y)*3/2
    if u1 <= pa:
        return Y
    else:
        return random_number_generation()

#Generation of 50000 random values of the distribution
random_values = [random_number_generation() for _ in range(100000)]

#Display of the PDF's approximation with the histogram of generated values
x = np.linspace(min(random_values),max(random_values),100)
y = [pdf_display(i) for i in x]
plt.hist(random_values, 100, density = True)                  #Display of the histogram generated from the random values
plt.plot(x, y, color = 'green')                                 #Display of the original PDF

plt.show()
```

### **Punto 2 b**

```python
import numpy as np
import random as r
import matplotlib.pyplot as plt

def function_display(x):
  if 0 <= x <= 1:
    return 1/4
  elif 1<x<=(np.sqrt(73)+11)/12:
    return 3*x - 11/4
  else:
    return 0

def random_function():
  seed1,seed2 = r.uniform(0,1),r.uniform(0,1)
  x_t = seed1*(np.sqrt(73)+11)/12
  if x_t <= 1:
    pa = 1/np.sqrt(73)
  else:
    pa = (12*x_t - 11)/np.sqrt(73)
  if seed2 <= pa:
    return x_t
  else:
    return random_function()

#Generation of a list with 50000 random values of an triangular distribution with a,b and c parameter
test_list = [random_function() for i in range(100000)]

#Graph display
x = list(np.linspace(min(test_list),max(test_list),100))
y = [function_display(i) for i in x]
plt.hist(test_list, 100, density = True)                       #Display for histogram from random data
plt.plot(x, y, color = 'red')                                 #Display for probability distribution function
plt.show()
```

### **Punto 2 c**

```python
import numpy as np
import random as r
import matplotlib.pyplot as plt

def function_display(x):
  if 0 <= x <= 1:
    return 3/4
  elif 1<x<=2:
    return 1/4
  else:
    return 0

def random_function():
  seed1,seed2 = r.uniform(0,1),r.uniform(0,1)
  x_t = 2*seed1
  if x_t <= 1:
    pa = 1
  else:
    pa = 1/3
  if seed2 <= pa:
    return x_t
  else:
    return random_function()

#Generation of a list with 50000 random values of an triangular distribution with a,b and c parameter
test_list = [random_function() for i in range(100000)]

#Graph display
x = list(np.linspace(min(test_list),max(test_list),100))
y = [function_display(i) for i in x]
plt.hist(test_list, 100, density = True)                       #Display for histogram from random data
plt.plot(x, y, color = 'red')                                 #Display for probability distribution function
plt.show()
```

```python
import numpy as np
import random as r
import matplotlib.pyplot as plt

def function_display(x):
  if 4<=x<=5:
    return (-x/2)+ (11/4)
  else:
    return (x/2)- (9/4)


def random_function():
  seed1,seed2 = r.uniform(0,1) , r.uniform(0,1)              #Random value between 0 and 1
  x_t = 2*seed1 + 4                                          #Get value to accept or reject
  if x_t<=5:                                                 #Process to get the not rejection probability
    pa = (-2*x_t + 11)/3
  else:
    pa = (2*x_t/3) - 3
  if seed2 <= pa:
    return x_t                                              #Value returned as it was not rejected
  else:
    return random_function()                                #Value rejected (Retry process to get a new one)

#Generation of a list with 50000 random values of distribution defined
test_list = [random_function() for i in range(100000)]

#Graph display
x = list(np.linspace(min(test_list),max(test_list),100))
y = [function_display(i) for i in x]
plt.hist(test_list, 100, density = True)                       #Display for histogram from random data
plt.plot(x, y, color = 'red')                                 #Display for probability distribution function

plt.show()
```
