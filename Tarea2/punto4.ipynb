Función para graficar variables generadas

```python
# Graficar histograma de las variables generadas
def show_variates(variates_list,num_bins,title):
  plt.hist(variates_list,bins=num_bins) # Entre mayor número 'bins', más preciso el histograma
  plt.xlabel("Valores")
  plt.ylabel("Frecuencia")
  plt.title(title)
  plt.grid(True)
  plt.show()
```

Función para calcular la media de una lista

```python
def get_mean(list):
    array = np.array(list)
    return np.mean(array)
```

# Punto 4
Desarrollar  un esquema general  para generar valores aleatorios para la distribución triangular  con pdf:



Genere 10 valores de la variable aleatoria, evalué la media muestral , y compare con la media de la distribución


Generación de variables

```python
np.random.seed(0)

def generator_1(rand_u):
  return 2+np.sqrt(4*rand_u)

def generator_2(rand_u):
  return 6-np.sqrt(12*(1-rand_u))

# Generando 10 variables
iterations = 10
variates_10 = []
for i in range(0,iterations):
  rand_u = np.random.uniform(0,1)
  if rand_u <= (1/4):
    variates_10.append(generator_1(rand_u))
  else:
    variates_10.append(generator_2(rand_u))

print("-"*40)
print(f"Variables generadas = {iterations}")
print(f"Promedio = {get_mean(variates_10)}")
print("-"*40)

# Generando un millón de variables
iterations = 1000000
variates_1M = []
for i in range(0,iterations):
  rand_u = np.random.uniform(0,1)
  if rand_u <= (1/4):
    variates_1M.append(generator_1(rand_u))
  else:
    variates_1M.append(generator_2(rand_u))

print(f"Variables generadas = {iterations}")
print(f"Promedio = {get_mean(variates_1M)}")
print("-"*40)
```

Reporte de resultados (gráficas)

```python
# GRAFICA DE LAS VARIABLES GENERADAS
show_variates(variates_10,50,"Punto 4 - generando 10 variables")
show_variates(variates_1M,50,"Punto 4 - generando un millón de variables")

# GRAFICA DE LA DISTRIBUCIÓN
def funcion1(x):
  return (1/2)*(x-2)

def funcion2(x):
  return (1/2)*(2-(x/3))

dominio_1 = np.linspace(2,3, 100)
dominio_2 = np.linspace(3,6, 100)

# Calcular los puntos para cada tramo
tramo_1 = funcion1(dominio_1)
tramo_2 = funcion2(dominio_2)

# Graficar los tramos
plt.plot(dominio_1, tramo_1, label="Tramo 1")
plt.plot(dominio_2, tramo_2, label="Tramo 2")

# Agregar etiquetas y título
plt.xlabel("x")
plt.ylabel("f(x)")
plt.ylim(0, 0.52)
plt.title("Distribución punto 4")
plt.legend()
plt.grid(True)  # Add gridlines for better readability (optional)
plt.show()
```

### Usando el algoritmo general del punto 1.a, generando un millón de variables

```python
np.random.seed(0)

# PARÁMETROS DE LA DISTRIBUCIÓN TRIANGULAR
a=2
b=3 # media
c=6

def generator_a_1(a,b,c,rand_u):
  return a + math.sqrt(rand_u*(c-a)*(b-a))

def generator_a_2(a,b,c,rand_u):
  return c - math.sqrt((1-rand_u)*(c-a)*(c-b))

iterations = 1000000
variates = []
for i in range(0,iterations):
  rand_u = np.random.uniform(0,1)
  if rand_u <= (b-a)/(c-a):
    variates.append(generator_a_1(a,b,c,rand_u))
  else:
    variates.append(generator_a_2(a,b,c,rand_u))
```

```python
# GRAFICA DE LAS VARIABLES GENERADAS
show_variates(variates,50,"Punto 4 - generando un millón de variables")

# GRAFICA DE LA DISTRIBUCIÓN
def funcion1(x):
  return (1/2)*(x-2)

def funcion2(x):
  return (1/2)*(2-(x/3))

dominio_1 = np.linspace(2,3, 100)
dominio_2 = np.linspace(3,6, 100)

# Calcular los puntos para cada tramo
tramo_1 = funcion1(dominio_1)
tramo_2 = funcion2(dominio_2)

# Graficar los tramos
plt.plot(dominio_1, tramo_1, label="Tramo 1")
plt.plot(dominio_2, tramo_2, label="Tramo 2")

# Agregar etiquetas y título
plt.xlabel("x")
plt.ylabel("f(x)")
plt.ylim(0, 0.52)
plt.title("Distribución punto 4")
plt.legend()
plt.grid(True)  # Add gridlines for better readability (optional)
plt.show()
```
